```python
Y=data.iloc[:,:1].values
```

```python
X=data.iloc[:,1:].values
```

```python
m=X.shape[0] #number of independant variables
n=X.shape[1] #number of dependant variables
```

```python
ones=np.ones((m,1))
```

```python
X=np.concatenate((ones,X),axis=1)
```

```python
def costfn_and_grad_descent(X,Y,alpha,max_iter):  #alpha is learning rate
   cost=np.zeros(max_iter)
   Theta=np.zeros(n+1) #initial theta vector
   h=np.dot(X, Theta) #hypothesis
   sgm= 1/(1+np.exp(-h)) #sigmoid function
   for i in range(0,max_iter):
      for j in range(0,n):
        Theta[j]=Theta[j]-(alpha/m)*(np.dot(sum(sgm-Y),X[:,j]))
      h=np.dot(X,Theta)
      cost[i]= -(1/m)*(np.sum((Y*(np.log(sgm)))+(1-Y)*(np.log(1-sgm))))
   return cost, Theta
```

```python
cost, Theta=costfn_and_grad_descent(X,Y,0.01,10)
```
