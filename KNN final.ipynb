```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
def find_neighbours(k, X_train, new_point):
  neighbour_arr=[]
  for i in range(len(X_train)):
    dist=np.sqrt(sum(np.square(X_train[i]-new_point))) #euclidean distance
    neighbour_arr.append([i,dist])
    neighbour_arr=sorted(neighbour_arr,key=lambda x:x[1])

  return print(np.array(neighbour_arr[0:k])), print('\n')
```
